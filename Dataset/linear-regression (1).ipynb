```python


import numpy as np 
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt 



import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))

```

### First we found a suitable dataset, imported it into our notebook, and assigned it as a Pandas dataframe:

```python
filename = '../input/indias-population-simple-linear-regression/population.csv'
df = pd.read_csv(filename)
```

### Viewing the top 5 rows, we can see how the data will be arranged:

```python
df.head(5)
```

### The `.describe()` method allows us to view the measures of central tendency of the dataframe at a glance:

```python
df.describe()
```

### The `.info()` method allows us to view all column names, their counts, and their data types supported within the data frame:

```python
df.info()
```

### Next, we separate the data into a training set and testing sets:

```python
x = df['Year'].values
y = df['Population'].values

x = x.reshape(-1,1)
y = y.reshape(-1,1)

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=.8, test_size=.2, random_state=17)
```

### Here we can see the data charted:

```python
plt.scatter(x_train, y_train, color='blue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India Over Time')
plt.show()
```

### Next, using a linear regression model, we predict the trend of the population data:

```python
lm = LinearRegression()
lm.fit(x_train,y_train)
y_predict =lm.predict(x_test)
```

### We then test the accuracy score, or how closely our data sticks to the predicted trend. We see our data was extremely consistent:

```python
print(f'Train Accuracy {round(lm.score(x_train,y_train)*100, 2)}%')
print(f'Test Accuracy {round(lm.score(x_test,y_test)*100, 2)}%')
```

### Lastly, we draw our linear regression line:

```python
plt.scatter(x_train, y_train, color='blue')
plt.plot(x_test,y_predict)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population of India Over Time')
plt.show()
```
